### Top 10 Total Boundaries = Fours + Sixes

```python
total_fours = all_season_batting_card.groupby('name')['fours'].sum().sort_values(ascending = False)
total_sixes = all_season_batting_card.groupby('name')['sixes'].sum().sort_values(ascending = False)

total_boundaries = pd.merge(pd.DataFrame(total_fours), pd.DataFrame(total_sixes), on = 'name', how = 'inner')
total_boundaries['total'] = total_boundaries['fours'] + total_boundaries['sixes']
top_10_boundaries = total_boundaries.sort_values(by = 'total', ascending = False)

plt.figure(figsize = (20, 6))
sns.set_theme(font_scale=3)
top_10_boundaries.iloc[:10, :]['total'].plot(kind = 'bar')
plt.xlabel('BATSMEN')
plt.ylabel('TOTAL BOUNDARIES')
plt.title('TOTAL BOUNDARIES SCORED IN ALL SEASONS')
plt.show()
```

- Even though CH Gayle has not scored that many fours, but the number of sixes compensate for that making him the batsman with highets count of boundaries. He is a real hard hitter of the ball!
- All the other batsmen are quite close in terms of total boundaries.

- It is more informative to calculate the strike rate of batsmen who have scored much more runs (regular batsmen)

### Total runs scored by teams across all seasons

```python
teams_total_runs = []

for season in np.unique(all_season_summary['season']):
    all_season_summary_season = all_season_summary[all_season_summary['season'] == season]
    runs_teams_home_season = []
    for team in unique_teams:
        runteam = all_season_summary_season[all_season_summary_season['home_team'] == team]
        runs_teams_home_season.append(list(runteam['home_runs']))
    
    runs_team_away_season = []
    for team in unique_teams:
        runteam = all_season_summary_season[all_season_summary_season['away_team'] == team]
        runs_team_away_season.append(list(runteam['away_runs']))
        
    for i in range(len(runs_teams_home_season)):
        runs_teams_home_season[i].extend(runs_team_away_season[i])
    
    teams_total_runs.append(runs_teams_home_season)
    
```
